```python
from langchain_google_genai import ChatGoogleGenerativeAI
from dotenv import load_dotenv

load_dotenv()

model1 = ChatGoogleGenerativeAI(
    model = "gemini-2.5-flash"
)
```

```python
from langchain_ollama import ChatOllama

model2 = ChatOllama(
    model = "llama3"
)
```

```python
from langchain_core.prompts import PromptTemplate

prompt1 = PromptTemplate(
    template="Write a punchy, 280-character tweet about {topic}. Use 2 emojis and one hashtag",
    input_variables=["topic"]
) 

prompt2 = PromptTemplate(
    template="Write a professional 200-character LinkedIn post about {topic}. Focus on productivity and career benefits. Use bullet points",
    input_variables=["topic"]
)

```

```python
from langchain_core.output_parsers import StrOutputParser

parser = StrOutputParser()
```

```python
tweetter_chain = prompt1 | model1 | parser
linkedin_chain = prompt2 | model2 | parser 
```

```python
from langchain_core.runnables import RunnableParallel

parallel_chain = RunnableParallel({
    "tweet": tweetter_chain,
    "linkedin": linkedin_chain
})
```

```python
final = parallel_chain.invoke({"topic": "AI in Healthcare"})
```
