```python
class Calc:
    def __init__(self, img: int = 0, real: int = 0):
        self.i = img
        self.r = real
        
    def __str__(self):
        return f"real: {self.r} + img: {self.i}"
        
        
    def __add__(self, cal: Calc) -> Calc:
        img = self.i + cal.i
        real = self.r + cal.r
        return Calc(img,real)
    
    def __mul__(self, cal: Calc) -> Calc:
        img = self.i * cal.i
        real = self.r * cal.r - img
        return Calc(0, real)
    
    def __or__(self, cal: Calc) -> Calc:
        return self + cal

        
```

```python
c1 = Calc(2,5)
c2 = Calc(3,2)
c3 = c1 + c2
print(c3)
```

```python
c4 = c1 | c2 | c3 | c3
print(c4)
```
