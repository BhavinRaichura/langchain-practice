# Minimulistic RAG Application

### core component of RAG application
- sementic search
- cosine similarity
- embedding models
- vector DB (used to store the embeddings for sementic search)


### flow
- storing flow
    -> documents 
    -> embedding models 
    -> convert to embeddings
    -> store embedding to vector db 
- retrival flow
    -> query text
    -> embedding model
    -> create embeddings
    -> similarity search with existing doc (vector DB automatically do this task using selected distance fucntion. ie: cosine-similarity, eclunius distance etc)
    -> returns the K similar docs or silimarity probability of different docs



```python
from langchain_ollama import OllamaEmbeddings

embedding_model = OllamaEmbeddings(
    model="nomic-embed-text",
    base_url="http://localhost:11434"
)
```

```python
documents = [
    "everybudy should be come in the office with formal dress",
    "office timing is 11AM to 8PM",
    "late comers/leave without prior notice will be need to give fine",
    "for late comers afer 11AM the fine will changed equivalent to the deduction of half full day salary",
    "for late comers afer 3PM the fine will changed equivalent to the deduction of full day salary"
]
embeded_docs = embedding_model.embed_documents(documents)
```

```python
print(str(embeded_docs))
```

```python
query = "how much fine i need to pay if i go late late office after 5PM"
embeded_query = embedding_model.embed_query(query)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# need to compare matrics with same dimentions
# sementic search
scores = cosine_similarity([embeded_query], embeded_docs)[0]
```

```python
print(scores)
```

```python
rag_search = sorted(list(enumerate(scores)), key=lambda x:x[1])
index, score = rag_search[-1]
```

```python
documents[index]
```
